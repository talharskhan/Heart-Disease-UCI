```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_disease = pd.read_csv('heart.csv')
```

```python
X = heart_disease.drop(columns = 'target', axis = 1)
y = heart_disease['target']
```

```python
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .20, stratify = y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
x_train_pred = model.predict(X_train)
train_acc = accuracy_score(x_train_pred, y_train)
print("Accuracy of train data is : ", train_acc)
```

```python
x_test_pred = model.predict(X_test)
test_acc = accuracy_score(x_test_pred, y_test)
print("Accuracy of test data : ", test_acc)
```

```python
# Predictive system
input_data = (57,0,0,120,354,0,1,163,1,0.6,2,0,2)
np_input_data = np.asarray(input_data)

reshaped_data = np_input_data.reshape(1, -1)
# print(reshaped_data)

prediction = model.predict(reshaped_data)
print(prediction)

if (prediction[0] == 1) :
    print("Person Have Heart Disease")
else : 
    print("Person Does Not Have Heart Disease")
```
